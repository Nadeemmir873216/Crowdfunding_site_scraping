## Extract contact details 

```python
import requests, re
from bs4 import BeautifulSoup
import pandas as pd
import os

test_website_list = website_list[:5]

contact_info=[]
def extract_contacts(url):
    try:
        response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
        response.raise_for_status()  # Raise an exception for bad status codes (4xx or 5xx)
        soup = BeautifulSoup(response.text, "html.parser")
        text = soup.get_text()

        # Extract emails
        emails = re.findall(r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b", text)

        # Extract Instagram handles and create full URLs
        instagram_handles = re.findall(r"instagram\.com/([A-Za-z0-9_.]+)/?", str(soup))
        instagram_links = ["https://www.instagram.com/" + handle for handle in instagram_handles]

        return {"emails": emails, "instagram": instagram_links}
    except requests.exceptions.RequestException as e:
        print(f"Error accessing {url}: {e}")
        return {"emails": [],'instagram': []}
    except Exception as e:
        print(f"An unexpected error occurred while processing {url}: {e}")
        return {"emails": [],'instagram': []}

for website in website_list:
    contact = extract_contacts(website)
    if contact is not None:
        contact_info.append(contact)
```

```python
all_insta = []
for s_info in contact_info:
    link_list = s_info['instagram']
    seen = set()
    clean_llist = []
    for link in link_list:
        if link not in seen:
            seen.add(link)
            clean_llist.append(link)
    all_insta.append(clean_llist)
        
```

```python
print(all_insta)
```

```python
final_all_insta_links = []
for link in all_insta:
    link = str(link)
    link = link.replace("[", "")
    link = link.replace("]", "")
    link = link.replace("'", "")
    # print(link)
    final_all_insta_links.append(link)
# print(all_insta)
```
